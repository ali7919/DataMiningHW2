```python
df[['Name','Global_Sales']].groupby('Name').sum().sort_values(by='Global_Sales', ascending=False).head(10)
```

```python
top_platform = df[['Platform', 'Global_Sales',]].groupby('Platform').sum().sort_values(by='Global_Sales',ascending=False)
top_platform.head(10).plot(kind='pie', subplots=True,title='Global Sales per Platform',autopct='%1.0f%%')

```
